# Prime Number Generator

<font color='yellow'>Project description:</font> Generate a list of prime numbers from a maximum number

<font color='yellow'>Languages used for this project:</font> Python

<font color='yellow'>Test cases:</font>
* None: Exception
* Not an int: Exception
* Provide an entry equal to 20: [False, False, True, True, False, True, False, True, False, False, False, True, False, True, False, False, False, True, False, True]

<font color='yellow'>Assumptions of the algorithm:</font>
* For a number to be prime, it must equal 2 or greater and cannot be divisible by any number other than 1 or itself
* All non-prime numbers are divisible by a prime number

<font color='yellow'>Wikipedia animation that simulates the algorithm's operation:</font>

![alt text](https://upload.wikimedia.org/wikipedia/commons/b/b9/Sieve_of_Eratosthenes_animation.gif)

```python
import math

class PrimeGenerator(object):
    def generate_primes(self, max_num):
        if max_num == None or (max_num%2) != 0.5 and (max_num%2) != 0:
            raise TypeError('invalid max_num')
        listTemp = [True] * max_num
        listTemp[0] = False
        if max_num >= 1:
            listTemp[1] = False
        if max_num >= 2:
            listTemp[2] = True
        for num in range(3, max_num):
            for i in range(2, num+1):
                if (num % i) == 0 and num != i:
                    listTemp[num] = False
        return listTemp
```

```python
# Solution test:
value = 100
array_primes = PrimeGenerator()
array_primes.generate_primes(max_num=value)
```
